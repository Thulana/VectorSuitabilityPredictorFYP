```python
#seperate traindata and test data here
X_train, X_test, Y_train, Y_test = train_test_split(weightedX,y_train ,test_size=0.2)
```

# Logistic Regression

```python
lrmodel = LogisticRegression()
lrmodel.fit(X_train,Y_train)
```

# Logistic Regression Cross validating the model

```python
scores = cross_val_score(lrmodel, X_train, Y_train)
print (scores)

prediction = lrmodel.predict(X_test)
print (prediction)
print (Y_test)
accuracy_score(Y_test, prediction)
```

# Random Forest Classifier

```python
randomForest = RandomForestClassifier(n_estimators=11,n_jobs=-1,random_state=0)
randomForest.fit(X_train,Y_train)
```

# Random Forest Classifier Cross validating the model

```python
scores = cross_val_score(randomForest, X_train, Y_train)
print (scores)
```

```python
prediction = randomForest.predict(X_test)
print (prediction)
print (Y_test)
accuracy_score(Y_test, prediction)
```

# Ensembled Bagging Decison Tree Classifier

```python
baggingClassifier = ensemble.BaggingClassifier(n_estimators=10,bootstrap_features=True,n_jobs=-1)
baggingClassifier.fit(X_train,Y_train)
```

# Ensembled Bagging Decison Tree Cross validating the model

```python
scores = cross_val_score(baggingClassifier, X_train, Y_train)
print (scores)
```

```python
prediction = baggingClassifier.predict(X_test)
print (prediction)
print (Y_test)
accuracy_score(Y_test, prediction)
```
